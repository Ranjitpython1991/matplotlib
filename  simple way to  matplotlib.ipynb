```python

import matplotlib.pyplot as plt 
```

# Line plot :

```python
# x-axis values
x = [1, 2, 3, 4,5,6,7]
  
# Y-axis values
y = [10, 5, 8, 4, 2,5,6]
  
# Function to plot
plt.plot(x,y)
  
# function to show the plot
plt.show()
```

# Bar plot 

```python
# x-axis values
x = [1, 2, 3, 4,5,6,7]
  
# Y-axis values
y = [10, 5, 8, 4, 2,5,6]
  
# Function to plot the bar
plt.bar(x,y)
  
# function to show the plot
plt.show()
```

#  hist

```python
# x-axis values
x = [1, 2, 3, 4,5,6,7]
  
# Y-axis values
y = [10, 5, 8, 4, 2,5,6]
  
plt.hist(y)
  
# Function to show the plot
plt.show()
```

# Scatter Plot :

```python
# x-axis values
x = [1, 2, 3, 4,5,6,7]
  
# Y-axis values
y = [10, 5, 8, 4, 2,5,6]
  
# Function to plot scatter
plt.scatter(x, y)

# lable
plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')
plt.title("Scatter Plot")
# function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# data to display on plots
x = [1, 2, 3, 4,5,6,7]
plt.pie(x)
# Title to the plot
plt.title("Pie chart")
plt.show()
```
